### Task 1:Calculate a square root

We can calculate the square root of a number using Newton's method [1,2].
To find the square root $z$ of a number $x$, we can iterate using the following equation. 

$$ z_{next} = z - \frac{z^2 - x} {2z} $$

```python
def sqrt(x):
    """
    A funtion to calculate the square root of a number x. 
    """
    # Initial guess for the square root z.
    z = x / 2
    #Loopn until we're happy with the accuracy.
    while abs(x - (z * z)) > 0.001:
        # Calculate a better guess for the square root.
        z -= (z*z -x) / (2 * z)
    # Return the (approximate) square root of x.     
    return z
```

### Task 3: STDEV.P vs STDEV.S

The standard deviation $STDEV.P$ and $STDEV.S$ 

$STDEVP$ assumes that this is the entire population where $STDEV.S$ represents a sample of the population. STDEV.S deals with smaller sample numbers then the STDEV.P which deals with numbers that are alot bigger. These test work in a way that if a very small sample is tested a large difference can be seen form the tests.
My understanding of the stdev.p and stdev.s is that the "p" takes the whole population data set (all the elements) and works out the true mean. Where the "s" takes a sample of the of the data set and works out the mean. STDEV.S uses the $n-1$ for it formulas where STDEV.P uses $n$ for its formulas

### Task 4: kNN (k Nearest Neighbours)

$kNN$ or K-nearest neighbours. This is one of the simplest classification algorithms and is one of the most used learning algorithms. This algorithm is used to get data points from different data classes and this will help make a new classification of a sample point.  

The Iris flower data set is a multivariate data set introduced by Ronald Fisher in 1936 as an axample of discriminant analysis. The same Ronald Fisher who invented Fishers exact test. The discriminant function is used on $Iris Versicolor$ and $Iris Virginica$ but the $Iris Setosa$ is different from the other two species.
Image below: https://en.wikipedia.org/wiki/Iris_flower_data_set

```python
%matplotlib inline
```

Eample of the Iris Dataset
Imformation and research on kNN - https://towardsdatascience.com/knn-using-scikit-learn-c6bed765be75

```python
print(__doc__)


import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from sklearn import datasets
from sklearn.decomposition import PCA

# import some data to play with
iris = datasets.load_iris()
X = iris.data[:, :2]  # we only take the first two features.
y = iris.target

x_min, x_max = X[:, 0].min() - .5, X[:, 0].max() + .5
y_min, y_max = X[:, 1].min() - .5, X[:, 1].max() + .5

plt.figure(2, figsize=(8, 6))
plt.clf()

# Plot the training points
plt.scatter(X[:, 0], X[:, 1], c=y, cmap=plt.cm.Set1,
            edgecolor='k')
plt.xlabel('Sepal length')
plt.ylabel('Sepal width')

plt.xlim(x_min, x_max)
plt.ylim(y_min, y_max)
plt.xticks(())
plt.yticks(())

# To getter a better understanding of interaction of the dimensions
# plot the first three PCA dimensions
fig = plt.figure(1, figsize=(8, 6))
ax = Axes3D(fig, elev=-150, azim=110)
X_reduced = PCA(n_components=3).fit_transform(iris.data)
ax.scatter(X_reduced[:, 0], X_reduced[:, 1], X_reduced[:, 2], c=y,
           cmap=plt.cm.Set1, edgecolor='k', s=40)
ax.set_title("First three PCA directions")
ax.set_xlabel("1st eigenvector")
ax.w_xaxis.set_ticklabels([])
ax.set_ylabel("2nd eigenvector")
ax.w_yaxis.set_ticklabels([])
ax.set_zlabel("3rd eigenvector")
ax.w_zaxis.set_ticklabels([])

plt.show()

```

The Iris Dataset https://scikit-learn.org/stable/auto_examples/datasets/plot_iris_dataset.html

This algorithm data set catalogues the size and dimensions of the three different types of Iris petals and sepal lengths. The images above displays the Sepal length, Sepal width, Petal length and Petal width and displays them plotted out within the graphs.
